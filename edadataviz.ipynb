## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.


```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

Next, let's drill down to each site visualize its detailed launch records.


### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

# Plot FlightNumber vs LaunchSite
sns.catplot(data=df, x="FlightNumber", y="LaunchSite", hue="Class", aspect=5)

# Add titles and labels for clarity
plt.title("Flight Number vs Launch Site by Landing Success", fontsize=20)
plt.xlabel("Flight Number", fontsize=15)
plt.ylabel("Launch Site", fontsize=15)
plt.show()
```

Now try to explain the patterns you found in the Flight Number vs. Launch Site scatter point plots.


1. Improved Reliability Over Time
There is a clear progression in landing success as the Flight Number increases. Earlier flights (lower numbers) show a higher concentration of Class 0 (Failures), while later flights are almost exclusively Class 1 (Successes). This illustrates SpaceX's rapid learning curve and technical refinement. Seaborn Visualization Documentation.
2. Launch Site Specialization
CCAFS SLC-40: This was the "workhorse" site for early missions. It handled the majority of the first 40 flights and consequently shows the most mixed landing outcomes.
KSC LC-39A: This site was integrated mid-way through the program (around Flight 30). It shows a very high success rate, as it became active once SpaceX had already stabilized its landing technology.
VAFB SLC-4E: Used less frequently and primarily for specific polar orbit missions. Similar to KSC, it shows improved success rates in later flight sequences.
3. Flight Volume Trends
The density of points increases at higher flight numbers, particularly at CCAFS SLC-40 and KSC LC-39A, reflecting the increased launch cadence (frequency) SpaceX achieved as the Falcon 9 program matured. IBM SpaceX Dataset Analysis.
4. Site-Specific Observations
No single site is inherently "safer" than the others; rather, the success rate is more highly correlated with the Flight Number (technological maturity) than the physical location of the launch.

### TASK 2: Visualize the relationship between Payload Mass and Launch Site

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Group by Orbit and calculate the mean of the Class column
df_orbit_success = df.groupby('Orbit')['Class'].mean().reset_index()

# 2. Sort values for a more organized chart (optional but recommended)
df_orbit_success = df_orbit_success.sort_values(by='Class', ascending=False)

# 3. Create the bar chart
sns.barplot(data=df_orbit_success, x='Orbit', y='Class', hue='Orbit', palette='viridis', legend=False)

# 4. Labeling the plot
plt.title('Success Rate by Orbit Type', fontsize=20)
plt.xlabel('Orbit Type', fontsize=15)
plt.ylabel('Success Rate (0 to 1)', fontsize=15)
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.


### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

# Plot FlightNumber vs Orbit
sns.catplot(data=df, x="FlightNumber", y="Orbit", hue="Class", aspect=5)

# Formatting the chart
plt.title("Flight Number vs Orbit Type by Landing Success", fontsize=20)
plt.xlabel("Flight Number", fontsize=15)
plt.ylabel("Orbit", fontsize=15)
plt.show()
```

You can observe that in the LEO orbit, success seems to be related to the number of flights. Conversely, in the GTO orbit, there appears to be no relationship between flight number and success.


### TASK  5: Visualize the relationship between Payload Mass and Orbit type


Similarly, we can plot the Payload Mass vs. Orbit scatter point charts to reveal the relationship between Payload Mass and Orbit type


```python
# Plot a scatter point chart with x axis to be Payload Mass and y axis to be the Orbit, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

# Plot Payload Mass vs Orbit
sns.catplot(data=df, x="PayloadMass", y="Orbit", hue="Class", aspect=5)

# Adding descriptive labels and title
plt.title("Payload Mass vs Orbit Type by Landing Success", fontsize=20)
plt.xlabel("Payload Mass (kg)", fontsize=15)
plt.ylabel("Orbit Type", fontsize=15)
plt.show()
```

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Helper function to extract year from the date
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year

# 2. Reset year list and extract years
year=[]
df['Year'] = Extract_year()

# 3. Calculate average success rate per year
df_yearly_success = df.groupby('Year')['Class'].mean().reset_index()

# 4. Plot the line chart
plt.figure(figsize=(12,6))
sns.lineplot(data=df_yearly_success, x="Year", y="Class")

# 5. Add labels and title
plt.title("Yearly Launch Success Trend", fontsize=20)
plt.xlabel("Year", fontsize=15)
plt.ylabel("Success Rate", fontsize=15)
plt.grid(True)
plt.show()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# HINT: Use get_dummies() function on the categorical columns
# 1. Select the features to be transformed
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]

# 2. Use pandas get_dummies to create dummy variables for categorical columns
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# 3. Cast all numeric columns to float64 as required by some lab versions
features_one_hot = features_one_hot.astype('float64')

# 4. Verify the results
print(features_one_hot.head())
print(f"Total number of columns: {features_one_hot.shape[1]}")
```
